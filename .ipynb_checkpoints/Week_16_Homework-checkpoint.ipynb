```python
# Dependencies and modules:

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import EditedNearestNeighbours

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

from sklearn import metrics
from sklearn.metrics import classification_report, roc_curve, roc_auc_score
from sklearn.model_selection import cross_val_score

import matplotlib.pyplot as plt

from sklearn.svm import SVC

# diabetes.csv file:

diabetes_path = "diabetes.csv"
diabetes_df = pd.read_csv(diabetes_path)
diabetes_df.head(1)
```

### 1. Perform combined over and undersampling on the diabetes dataset (use SMOTEENN). Explain how combined sampling works.

```python
# create my testing classes
X = diabetes_df.drop('Outcome', axis=1)
y = diabetes_df['Outcome']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 42, stratify = y)

#find number of target variable to help guide resampling
y.value_counts()
```

```python
# define SMOTEENN
resample = SMOTEENN(enn=EditedNearestNeighbours(sampling_strategy='all'), random_state=42)
X_res, y_res = resample.fit_resample(X_train,y_train)
y_res.value_counts()
```

Combined sampling is widely considered the best approach to balancing your data if there is a disparity between two distince classes. It uses a COMBINATION of undersampling the majority with oversampling the minority to bring the data as close as possible to representing reality. SMOTEENN is SMOTE + ENN, or Synthetic Minority Oversampling TEchnique, which uses K-NN to find random points along lines between nearest neighbors of randomly selected points from the data to artificially create new points representing the minority, and Edited Nearest Neighbors, which uses K-NN to prune the data points from the majority class that least match their neighbors.

### 2. Perform logistic regression with the new data from question 1. 

```python
# I have to first instantiate my classifier:
sc = StandardScaler()
lrc = LogisticRegression()

# fit my model to the classifier:
lr_model = lrc.fit(X_res, y_res)

# find my predicted y values:
y_pred = lr_model.predict(X_res)
print(classification_report(y_res, y_pred))
```

### 2.a.  Comment on the performance of combined sampling vs the other approaches we have used for the diabetes dataset.

The combined sampling approach far outstrips the performance of prior approaches we have used on this dataset. KNN gave an accuracy in the mid 70% range, Logistic Regression with scaling gave a similar accuracy but improved precision and recall, and Logistic Regression with SMOTE pushed values to high 70% to low 80%s. SMOTEENN with scaling pushed the accuracy, precision and recall above 90%. This is a clear 10% improvement margin.

### 4. What is outlier detection? Why is it useful? What methods can you use for outlier detection?

Outlier detection is any algorithm that can identify and/or remove data points that are significantly removed from other 'normal' data points or trends. Because outliers are usually caused by calculation errors or aberrant events, including outliers in your data seperates your model from reality. Thus, removing outliers makes better data. But first you have to detect them. Methods pythonistas have at their disposal include:
* creating box plots that will let you visually identify any outliers for investigation 

* the isolation forest package in sklearn, an unsupervised machine learning technique that randomly selects data columns and values and models them in a way that can spot anomalies 

* One-Class SVM uses support vector algorithms that capture the density of the majority class, then classifies points on the extremes of the density function as outliers

* Local Outlier Factor uses the nearest-neighbor idea to calculate neighbors around any particular point and give point an outlier liklihood value based on how...lonely it's neighborhood is in relation to the other points in the dataset

* Minimum Covariance Determinant can be used on data with a normal or Gaussian distribution by simply defining an ellipse area that covers most of the data and anything outside of it is an outlier.
